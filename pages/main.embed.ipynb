> **Tip**
>
> -   In general, it is advisable to improve your prompting strategy by
>     **iterating**.
>
> -   Start with a simple prompt, see what the model generates, and then
>     refine. Repeat this process until you are satisfied with the
>     results.
>
> -   Once you have a good prompt, you can use it in a new chat
>     session - in this way, the “failed” attempts will not be part of
>     the context.
